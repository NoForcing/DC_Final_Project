```python
# scinder marque_annee en deux colonnes marque et modele 

# Utilisation de apply avec une fonction lambda pour scinder les chaînes
df['marque'] = df['marque_annee'].apply(lambda x: x.split(' ', 1)[0])
df['annee'] = df['marque_annee'].apply(lambda x: x.split(' ', 1)[1])
```

```python
# Séparation de la colonne marque en deux colonnes marque et modèle
df[['marque', 'modele']] = df['marque'].str.rsplit(' ', n=1, expand=True)
```

```python
# Repositionner les colonnes marque et annee comme premières
colonnes_ordonnees = ['marque', 'annee'] + [col for col in df.columns if col not in ['marque', 'annee']]
df = df[colonnes_ordonnees]
```

```python
# Repositionner les colonnes marque et modèle comme premières
colonnes_ordonnees = ['marque', 'modele','annee'] + [col for col in df.columns if col not in ['marque', 'annee']]
df = df[colonnes_ordonnees]
```

```python
df.drop("marque_annee", axis =1, inplace = True)
```

```python
# Supprimer les colonnes dupliquées en gardant la première occurrence
df = df.loc[:, ~df.columns.duplicated()]
```
